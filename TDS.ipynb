```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
train_data = pd.read_csv('fraudTrain.csv')
test_data = pd.read_csv('fraudTest.csv')
train_data.head()
```

```python
test_data.head()
```

```python
# Check for missing values
print(train_data.isnull().sum())

# Split the training data into features and target variable
X_train = train_data.drop('is_fraud', axis=1)  # Features
y_train = train_data['is_fraud']                # Target variable
```

```python
# Check the data types of the training dataset
print(train_data.dtypes)
```

```python
# Convert 'trans_date_trans_time' to datetime
train_data['trans_date_trans_time'] = pd.to_datetime(train_data['trans_date_trans_time'])

# Extract useful features from the datetime column
train_data['transaction_hour'] = train_data['trans_date_trans_time'].dt.hour
train_data['transaction_day'] = train_data['trans_date_trans_time'].dt.day
train_data['transaction_month'] = train_data['trans_date_trans_time'].dt.month
train_data['transaction_year'] = train_data['trans_date_trans_time'].dt.year
train_data['transaction_weekday'] = train_data['trans_date_trans_time'].dt.weekday  # 0=Monday, 6=Sunday
```

```python
train_data = train_data.drop(columns=['Unnamed: 0', 'trans_date_trans_time', 'trans_num', 'first', 'last', 'dob'])
```

```python
# Convert categorical columns to numeric using one-hot encoding
categorical_cols = ['merchant', 'category', 'gender', 'street', 'city', 'state', 'job']
train_data = pd.get_dummies(train_data, columns=categorical_cols, drop_first=True)
```

```python
# Split the data into features and target variable
X_train = train_data.drop('is_fraud', axis=1)  # Features
y_train = train_data['is_fraud']                # Target variable
```

```python
# Split the data into training and validation sets
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Downsample the dataset
X_train_sample, _, y_train_sample, _ = train_test_split(X_train_split, y_train_split, test_size=0.9, random_state=42)

# Create the Logistic Regression model
log_model = LogisticRegression(max_iter=1000)

# Fit the model to the downsampled data
log_model.fit(X_train_sample, y_train_sample)

```

```python
# Make predictions on the validation set for Logistic Regression
y_val_pred_log = log_model.predict(X_val)
accuracy_log = accuracy_score(y_val, y_val_pred_log)
print(f"Logistic Regression Validation Accuracy: {accuracy_log * 100:.2f}%")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X_train_sample, _, y_train_sample, _ = train_test_split(X_train_split, y_train_split, test_size=0.9, random_state=42)

tree_model = DecisionTreeClassifier(random_state=42)

tree_model.fit(X_train_sample, y_train_sample)


```

```python
# Make predictions on the validation set for Decision Tree
y_val_pred_tree = tree_model.predict(X_val)
accuracy_tree = accuracy_score(y_val, y_val_pred_tree)
print(f"Decision Tree Validation Accuracy: {accuracy_tree * 100:.2f}%")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train_sample, _, y_train_sample, _ = train_test_split(X_train_split, y_train_split, test_size=0.9, random_state=42)

# Create the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model to the downsampled data
rf_model.fit(X_train_sample, y_train_sample)

```

```python
# Make predictions on the validation set for Random Forest
y_val_pred_rf = rf_model.predict(X_val)
accuracy_rf = accuracy_score(y_val, y_val_pred_rf)
print(f"Random Forest Validation Accuracy: {accuracy_rf * 100:.2f}%")

```
